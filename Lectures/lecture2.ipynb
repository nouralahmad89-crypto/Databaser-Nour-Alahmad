Dataintegritet = databasen ska innehålla korrekt och giltig data.


Det betyder att databasen försöker skydda sig från:

- fel
- motsägelser
- trasiga relationer
- ogiltiga värden

Exempel:

1. man kan inte lägga in en order med ett kund-ID som inte finns
2. ett pris kan inte vara negativt
3. en primärnyckel får inte dupliceras

Det skyddas ofta med:

- PRIMARY KEY
- FOREIGN KEY
- NOT NULL
- CHECK constraints

### **Database Normalization**

Handlar om hur man strukturerar databasen.

Målet är att:

- minska duplicerad data
- organisera tabeller smart
- undvika problem när data ändras

Exempel:

dela upp Students och Courses i olika tabeller istället för allt i en stor tabell.

### **SQL Constraints**

Constraints = regler som skyddar dataintegritet i databasen.

Om man försöker lägga in fel data → databasen ger felmeddelande.

| Constraint | Funktion | Exempel |
|---|---|---|
| NOT NULL | Kolumnen måste ha värde | `Name VARCHAR(50) NOT NULL` |
| UNIQUE | Inga dubbletter tillåtna | `Email VARCHAR(100) UNIQUE` |
| PRIMARY KEY | Unik identifierare för varje rad | `Id INT PRIMARY KEY` |
| FOREIGN KEY | Kopplar tabeller och skyddar relationer | `FOREIGN KEY (country_id) REFERENCES country(country_id)` |
| CHECK | Kontrollerar villkor | `CHECK (Price >= 0)` |
| DEFAULT | Sätter standardvärde automatiskt | `Status DEFAULT 'Active'` |

---

### Named Constraints

```sql
CONSTRAINT PK_city PRIMARY KEY (city_id)
CONSTRAINT FK_city_country FOREIGN KEY (country_id)
REFERENCES country(country_id)
```

Vanliga prefix:
- `PK_` = Primary Key
- `FK_` = Foreign Key
- `UQ_` = Unique
- `CHK_` = Check
- `DF_` = Default

---

### Exempel

```sql
CREATE TABLE Users
(
    Id INT IDENTITY,

    Name VARCHAR(50) NOT NULL,

    Email VARCHAR(100) UNIQUE,

    Age INT CHECK (Age >= 18),

    Status VARCHAR(20) DEFAULT 'Active',

    CONSTRAINT PK_Users
        PRIMARY KEY (Id)
)
```

### **SQL Triggers**

Trigger = automatisk kod som körs när något händer i databasen.

Triggers kan köras vid:
- INSERT
- UPDATE
- DELETE

---

#### Enkel Syntax

```sql
CREATE TRIGGER trg_UserInsert
ON Users
AFTER INSERT
AS
BEGIN

    PRINT 'Ny användare skapad'

END
```

---

#### AFTER INSERT

```sql
AFTER INSERT
```

Kör triggern efter att en rad lagts till.

---

#### AFTER UPDATE

```sql
AFTER UPDATE
```

Kör triggern efter att data ändrats.

---

#### AFTER DELETE

```sql
AFTER DELETE
```

Kör triggern efter att en rad tagits bort.

---

#### INSERTED-tabellen

`INSERTED` är en temporär tabell som innehåller nya rader.

#### Exempel

```sql
INSERT INTO Users
VALUES (1, 'Ali')
```

`INSERTED` innehåller:

| Id | Name |
|---|---|
| 1 | Ali |

---

#### DELETED-tabellen

`DELETED` innehåller gamla rader som tagits bort eller ändrats.

## Exempel

```sql
DELETE FROM Users
WHERE Id = 1
```

`DELETED` innehåller:

| Id | Name |
|---|---|
| 1 | Ali |

---

#### UPDATE använder båda

Vid UPDATE:

- `DELETED` = gamla värden
- `INSERTED` = nya värden

#### Exempel

```sql
UPDATE Users
SET Name = 'Ahmed'
WHERE Id = 1
```

### DELETED

| Id | Name |
|---|---|
| 1 | Ali |

### INSERTED

| Id | Name |
|---|---|
| 1 | Ahmed |

---

# AFTER vs INSTEAD OF

| Typ | Funktion |
|---|---|
| AFTER | Körs efter operationen |
| INSTEAD OF | Ersätter operationen |

---

#### Exempel INSTEAD OF

```sql
CREATE TRIGGER trg_BlockInsert
ON Users
INSTEAD OF INSERT
AS
BEGIN

    PRINT 'Insert blockerad'

END
```

Ingen rad sparas automatiskt.
