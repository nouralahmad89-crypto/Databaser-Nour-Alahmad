En databas är ett program som lagrar information och gör det möjligt att snabbt söka och manipulera data.

Databas = smart lagring av data
Du kan:
- spara data
- söka snabbt
- göra avancerade frågor

Det är inte bara lagring – det handlar om att jobba med data effektivt.

SQL = språket du använder för att prata med databasen

**Kursens mål:**

Du ska kunna:

- skriva SQL-frågor (queries)
- använda databaser för:
- dataanalys
- skapa features (för ML!)

Superviktig koppling till AI: Du använder databaser för att:
- plocka ut rätt data
- bygga dataset till modelle

### **SQL (Structured Query Language)**

#### Olika databassystem

Vi kommer jobba med två:

1. MSSQL (Microsoft SQL Server)
- Vanlig i företag/institutioner
- Bra prestanda
- Lite annorlunda syntax
2. PostgreSQL
- Vanlig i moderna projekt/startups
- Följer standard bättre
- Mer “ren” SQL

relationsdatabaser : Det finns kopplingar mellan kolumner i tabeller . Varje kolumn kan ha en relation till en annan. En **databashanterare** är mjukvara för att söka och manipulera data .Exempel:PostgreSQL ,MSSQL.

Relationsdatabas = 
- data i tabeller
- kolumner har relationer
du kan:
- koppla tabeller
- skapa regler
- göra avancerade queries

- Relationsdatabaser = kopplad data
- SQL = språk för att jobba med data
- Queries = frågor du skickar
- Resultat = data tillbaka
- Avancerade queries = kraftfull analys

Flödet:`VS Code → SQL query → SQL Server → Database → Result tillbaka`

1. SSMS (SQL Server Management Studio)

Detta är databas-verktyget

Du använder det för:

- ✔ installera / restore .bak
- ✔ se alla databaser
- ✔ se tabeller, struktur
- ✔ administrera (users, permissions)
- ✔ felsöka

Tänk: “detta är kontrollrummet för databasen”

Du BEHÖVER SSMS när:
- du har en .bak fil
- du ska skapa databaser
- du ska fixa problem / permissions

2. VS Code + mssql extension

Detta är kodverktyget

Du använder det för:

- ✔ skriva SQL-kod
- ✔ spara dina lösningar
- ✔ jobba i projekt
- ✔ version control (Git)

Tänk: “detta är där jag programmerar”

Så här jobbar du smart
**SSMS**
-  kolla vad finns i databasen
-  förstå tabeller
**VS Code**
- skriva lösningar
- köra queries

Sammanfattning av hela SELECT-delen:
- SELECT → hämta data
- WHERE → filtrera
- AND/OR → logik
- IN → lista
- BETWEEN → intervall
- LIKE → mönster
- ORDER BY → sortera
- DISTINCT → unika värden
- AS → alias
- UNION → kombinera
- CASE → logik
Vad SQL mer kan göra?
- INSERT (lägga till)
- UPDATE (ändra)
- DELETE (ta bort)
- skapa tabeller
- skapa views
- göra aggregation (sum, avg, etc)
- skriva funktioner
- skriva stored procedures

Det finns två skolor:

1. Lägg logik i databasen
färre system
2. Lägg logik i kod (t.ex. Python)
lättare att underhålla

Ingen rätt/fel – beror på projekt

- INSERT → lägga till
- SELECT INTO → skapa tabell
- INSERT + SELECT → kopiera data
- UPDATE → ändra
- DELETE → ta bort
- TRUNCATE → snabb rensning
- DROP → ta bort tabell
- SQL Injection → säkerhet
- Transaktioner → kontroll
